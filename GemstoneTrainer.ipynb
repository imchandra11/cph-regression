```python
# Direct Inference Example (Without API)
# This demonstrates how to use the trained regression model for price predictions

import joblib
import onnxruntime as ort
import numpy as np
import pandas as pd
from pathlib import Path

# Setup paths
MODELS_DIR = Path("GemstonePricePriction/models")

print("=" * 80)
print("Loading Models for Inference")
print("=" * 80)

# Load preprocessor
print("\n1. Loading preprocessor...")
preprocessor = joblib.load(MODELS_DIR / "preprocessor.joblib")
print("   ✓ Preprocessor loaded")

# Load ONNX model
print("\n2. Loading ONNX model...")
onnx_model_path = MODELS_DIR / "gemstone_price_model.onnx"
session = ort.InferenceSession(str(onnx_model_path))
print("   ✓ ONNX model loaded")

print("\n" + "=" * 80)
print("Running Inference Examples")
print("=" * 80)

# Define feature columns (must match training config)
categorical_cols = ['cut', 'color', 'clarity']
numeric_cols = ['carat', 'depth', 'table', 'x', 'y', 'z']
feature_cols = categorical_cols + numeric_cols

# Example 1: Sample gemstone prediction
print("\n" + "-" * 80)
print("Example 1: Sample Gemstone Price Prediction")
print("-" * 80)

sample_data = pd.DataFrame({
    'cut': ['Ideal'],
    'color': ['G'],
    'clarity': ['VS1'],
    'carat': [1.0],
    'depth': [61.5],
    'table': [57.0],
    'x': [6.5],
    'y': [6.55],
    'z': [4.0],
})

print("\nInput Features:")
print(sample_data[feature_cols].to_string(index=False))

# Transform data
transformed = preprocessor.transform(sample_data[feature_cols])
print(f"\nTransformed shape: {transformed.shape}")

# Predict
input_name = session.get_inputs()[0].name
output = session.run(None, {input_name: transformed.astype(np.float32)})
predicted_price = float(output[0][0][0])  # Regression: single value output

print("\n" + "=" * 80)
print("PREDICTION RESULT")
print("=" * 80)
print(f"\nPredicted Price: ${predicted_price:,.2f}")

# Example 2: Multiple predictions from dataset
print("\n\n" + "=" * 80)
print("Example 2: Predicting on Real Dataset Samples")
print("=" * 80)

# Load actual dataset
dataset_path = Path("GemstonePricePriction/data/gemstone.csv")
if dataset_path.exists():
    df = pd.read_csv(dataset_path)
    
    # Take first 5 samples
    test_samples = df.head(5)
    
    print(f"\nTesting on {len(test_samples)} samples from dataset:")
    print("-" * 80)
    
    predictions = []
    actual_prices = []
    
    for idx, row in test_samples.iterrows():
        # Prepare input
        input_data = pd.DataFrame([row[feature_cols]])
        
        # Transform and predict
        transformed = preprocessor.transform(input_data[feature_cols])
        output = session.run(None, {input_name: transformed.astype(np.float32)})
        predicted_price = float(output[0][0][0])
        actual_price = float(row['price'])
        
        predictions.append(predicted_price)
        actual_prices.append(actual_price)
        
        # Calculate error
        error = abs(predicted_price - actual_price)
        error_percent = (error / actual_price) * 100 if actual_price > 0 else 0
        
        print(f"\nSample {idx + 1}:")
        print(f"  Carat: {row['carat']:.2f}, Cut: {row['cut']}, Color: {row['color']}, Clarity: {row['clarity']}")
        print(f"  Dimensions: {row['x']:.2f} x {row['y']:.2f} x {row['z']:.2f} mm")
        print(f"  Actual Price:    ${actual_price:,.2f}")
        print(f"  Predicted Price: ${predicted_price:,.2f}")
        print(f"  Error:           ${error:,.2f} ({error_percent:.2f}%)")
    
    # Calculate overall metrics
    predictions = np.array(predictions)
    actual_prices = np.array(actual_prices)
    
    mae = np.mean(np.abs(predictions - actual_prices))
    mse = np.mean((predictions - actual_prices) ** 2)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((actual_prices - predictions) / actual_prices)) * 100
    
    print("\n" + "=" * 80)
    print("SUMMARY STATISTICS")
    print("=" * 80)
    print(f"Mean Absolute Error (MAE):     ${mae:,.2f}")
    print(f"Root Mean Squared Error (RMSE): ${rmse:,.2f}")
    print(f"Mean Absolute % Error (MAPE):   {mape:.2f}%")

# Example 3: Price comparison for different gemstones
print("\n\n" + "=" * 80)
print("Example 3: Price Comparison - Same Carat, Different Qualities")
print("=" * 80)

# Compare different quality levels for same carat weight
base_carat = 1.0
base_depth = 61.5
base_table = 57.0
base_x, base_y, base_z = 6.5, 6.55, 4.0

quality_scenarios = [
    {'cut': 'Ideal', 'color': 'D', 'clarity': 'IF', 'label': 'Best Quality'},
    {'cut': 'Very Good', 'color': 'G', 'clarity': 'VS1', 'label': 'High Quality'},
    {'cut': 'Good', 'color': 'I', 'clarity': 'VS2', 'label': 'Medium Quality'},
    {'cut': 'Fair', 'color': 'J', 'clarity': 'SI1', 'label': 'Lower Quality'},
]

print(f"\nComparing prices for {base_carat} carat gemstones with different qualities:")
print("-" * 80)

for scenario in quality_scenarios:
    test_data = pd.DataFrame({
        'cut': [scenario['cut']],
        'color': [scenario['color']],
        'clarity': [scenario['clarity']],
        'carat': [base_carat],
        'depth': [base_depth],
        'table': [base_table],
        'x': [base_x],
        'y': [base_y],
        'z': [base_z],
    })
    
    transformed = preprocessor.transform(test_data[feature_cols])
    output = session.run(None, {input_name: transformed.astype(np.float32)})
    predicted_price = float(output[0][0][0])
    
    print(f"{scenario['label']:20s}: Cut={scenario['cut']:12s}, Color={scenario['color']}, Clarity={scenario['clarity']:5s} → ${predicted_price:>10,.2f}")

print("\n" + "=" * 80)
print("Inference Complete!")
print("=" * 80)
```
